```python
import os
import matplotlib.pyplot as plt
```

```python
OPENSLIDE_PATH = 'C:/openslide-bin-4.0.0.6-windows-x64/bin'

import os
if hasattr(os, 'add_dll_directory'):
    with os.add_dll_directory(OPENSLIDE_PATH):
        import openslide
else:
    import openslide
```

```python
# Image
WSI_path = 'tumors/tumor_002.tif'
```

```python
# Mask
WSI_mask_path = 'tumors/tumor_002_mask.tif'
```

```python
# Annotation
WSI_annot = 'tumors/tumor_002.xml'
```

```python
img = openslide.OpenSlide(WSI_path)

levels = img.level_count

# Printing each zoom level of images in tiff file and their dimensions
for i in range(levels):
    lev_dim = img.level_dimensions[i]
    print(f"Level {i} - {lev_dim}")
```

```python
mask = openslide.OpenSlide(WSI_path)

levels = mask.level_count

# Printing each zoom level of images in tiff file and their dimensions
for i in range(levels):
    lev_dim = mask.level_dimensions[i]
    print(f"Level {i} - {lev_dim}")
```

```python
# Looking at whole slide
img_total = img.read_region((0, 0), 7, (764, 1716)).convert("RGB")

plt.figure(figsize=(8, 8))
plt.imshow(img_total)
```

```python
# Looking at whole mask
mask_total = mask.read_region((0, 0), 7, (764, 1716)).convert("RGB")

plt.figure(figsize=(8, 8))
plt.imshow(mask_total)
```

```python
# Looking at a patch of image
img_patch1 = img.read_region((40000, 110000), 0, (5000, 5000)).convert("RGB")

plt.figure(figsize=(8, 8))
plt.imshow(img_patch1)
```
